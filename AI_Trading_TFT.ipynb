```python
import pandas as pd

df = pd.read_csv('candlestick_features.csv')
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

# مشخص کردن ستون‌های ویژگی (feature)
feature_cols = df.columns.tolist()
feature_cols.remove('Close')  # ستون هدف رو حذف می‌کنیم (بعداً ازش برای پیش‌بینی استفاده می‌کنیم)

scaler = StandardScaler()
df[feature_cols] = scaler.fit_transform(df[feature_cols])
```

```python
df['time_idx'] = range(len(df))  # به هر کندل یه شماره زمانی میدیم
df['series_id'] = 'all'  # چون فقط یه سری دیتا داریم
```

```python
df[['time_idx', 'series_id'] + feature_cols + ['Close']].head()
```

```python
from pytorch_forecasting import TimeSeriesDataSet

# ستون هدف ما
target_col = "Close"

# ساخت دیتاست
dataset = TimeSeriesDataSet(
    df,
    time_idx="time_idx",
    target=target_col,
    group_ids=["series_id"],
    max_encoder_length=10,
    max_prediction_length=1,
    time_varying_known_reals=["time_idx"],
    time_varying_unknown_reals=feature_cols + [target_col],
)
```
